```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Importing data from Zenodo

We are using here data that we directly download from a Zenodo [repository](https://zenodo.org/record/1213207#.XPEnZ9P7TOQ), and which provides the data for [this paper](https://www.mdpi.com/2071-1050/10/4/1134/htm).

First we give the exact address of the dataset:

```python
zenodo_address = 'https://zenodo.org/record/1213207/files/Geneva_Eur11_TDY_2010_2039.txt'
```

Then we import the paper directly from Zenodo:

```python
data = pd.read_csv(zenodo_address,sep='\s+', skiprows = [0,1,2])
```

And we reproduce one of the plots:

```python
fig, ax = plt.subplots()
plt.plot(data.groupby('dm').mean().Ta)
ax.set_xlabel('Month')
ax.set_ylabel('Air temperature (CÂ°)')
plt.show()
```
